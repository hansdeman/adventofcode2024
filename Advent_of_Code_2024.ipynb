```python
with open("/content/drive/MyDrive/Vakanties, vrije tijd en recepten/Vrije Tijd/advent 2024/day2real.txt") as f:
    h = f.read()
    h = h.split("\n")
    h = [k.split(' ') for k in h]
    h = [[int(k) for k in j] for j in h]
print(h)
##part 1
safe = 0
direct = []
for k in h:
  if k[1] > k[0]:
    direct.append(1)
  else:
    direct.append(-1)

one_error_list = []
for k in range(len(h)):
  is_safe = 1
  for j in range(len(h[k])-1):
    if abs (h[k][j] - h[k][j+1]) > 3 or direct[k]*h[k][j] >= direct[k]*h[k][j+1]:
      is_safe = 0
      new = h[k].copy()
      one_error_list.append(new)
      break
  safe += is_safe
print(safe)
print(one_error_list)
##part 2

def safe_one_error(k):
  if k[1] > k[0]:
    direct = 1
  else:
    direct = -1

  is_safe = 1
  for j in range(len(k)-1):
    if abs (k[j] - k[j+1]) > 3 or direct*k[j] >= direct*k[j+1]:
      is_safe = 0
  return is_safe


safe_list = []
for k in one_error_list:
  is_safe = 0
  safe_list.append(0)
  for j in range(len(k)):
    temp = k.copy()
    temp.pop(j)
    is_safe = safe_one_error(temp)
    if is_safe == 1:
      safe += is_safe
      safe_list[-1] = 1
      break

print(safe_list)
print(safe)




```

```python
with open("/content/drive/MyDrive/Vakanties, vrije tijd en recepten/Vrije Tijd/advent 2024/day4.txt") as f:
    h = f.read()
    h = h.split("\n")

#part1

def transpose_strings(list_of_strings):
    """Transposes a list of strings as if they were a matrix."""
    max_len = len(list_of_strings)
    return [''.join(s[i] for s in list_of_strings) for i in range(max_len)]

v = transpose_strings(h)
print(h)
print(v)

def get_diagonals(matrix):
    """Prints all diagonals and anti-diagonals of a given matrix."""

    rows = len(matrix)
    cols = len(matrix[0])

    diagonals = []
    # Print diagonals (top-left to bottom-right)
    for i in range(rows):
        diagonal = ''
        row, col = i, 0
        while row >= 0 and col < cols:
            diagonal += matrix[row][col]
            row -= 1
            col += 1
        diagonals.append(diagonal)

    for j in range(1, cols):
        diagonal = ''
        row, col = rows - 1, j
        while row >= 0 and col < cols:
            diagonal += matrix[row][col]
            row -= 1
            col += 1
        diagonals.append(diagonal)

    # Print anti-diagonals (bottom-left to top-right)
    anti_diagonals = []
    for i in range(rows):
        anti_diagonal = ''
        row, col = i, cols - 1
        while row >= 0 and col >= 0:
            anti_diagonal += matrix[row][col]
            row -= 1
            col -= 1
        diagonals.append(anti_diagonal)

    for j in range(cols - 2, -1, -1):
        anti_diagonal = ''
        row, col = rows - 1, j
        while row >= 0 and col >= 0:
            anti_diagonal += matrix[row][col]
            row -= 1
            col -= 1
        diagonals.append(anti_diagonal)
    return diagonals

d = get_diagonals(h)
a = h+v+d
print(a)

matchingpattern = "SAMX"
test = []
for i in a:
  test.append(len(re.findall("XMAS", i))+len(re.findall("SAMX", i)))

count = len(test)
print(test)
print(sum(test))

#  part2
diag_matches = []
anti_matches = []


for i in range(len(h)-2):
  for j in range(len(h[0])-2):
    if h[i][j] == "M":
      if h[i+1][j+1] == "A":
        if h[i+2][j+2] == "S":
          diag_matches.append([i+1,j+1])
    if h[i][j] == "S":
      if h[i+1][j+1] == "A":
        if h[i+2][j+2] == "M":
          diag_matches.append([i+1,j+1])

for i in range(2,len(h)):
  for j in range(len(h[0])-2):
    if h[i][j] == "M":
      if h[i-1][j+1] == "A":
        if h[i-2][j+2] == "S":
          anti_matches.append([i-1,j+1])
    if h[i][j] == "S":
      if h[i-1][j+1] == "A":
        if h[i-2][j+2] == "M":
          anti_matches.append([i-1,j+1])


print(diag_matches)
print(anti_matches)
matches = 0

for i in diag_matches:
  if i in anti_matches:
    matches += 1

print(matches)
```
